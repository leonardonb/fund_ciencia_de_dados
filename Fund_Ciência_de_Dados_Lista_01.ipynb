**8) Apresente os gráficos das funções de Besel e da função de erro Gaussiana**

```python
from scipy.special import jv, erf

x = np.linspace(0, 10, 100)

y_bessel = jv(0, x)
y_erf = erf(x)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y_bessel)
plt.title("Função de Bessel de ordem 0")
plt.xlabel("x")
plt.ylabel("J₀(x)")

plt.subplot(1, 2, 2)
plt.plot(x, y_erf)
plt.title("Função de Erro Gaussiana")
plt.xlabel("x")
plt.ylabel("erf(x)")

plt.tight_layout()
plt.show()
```

**10) Apresente os gráficos dos polinômios de Legendre**

```python
from scipy.special import legendre

x = np.linspace(-1, 1, 100)

for n in range(10):
    Pn = legendre(n)
    y = Pn(x)
    plt.plot(x, y, label=f'P_{n}(x)')

plt.xlabel('x')
plt.ylabel('Pn(x)')
plt.title('Polinômios de Legendre')
plt.legend()
plt.grid(True)
plt.show()
```

**11) Apresente os gráficos das funções PDF e CDF (CDF é a integral de PDF (Função de Densidade de Probabilidade) e PDF é a derivada de CDF).**

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-3, 3, 100)
pdf = norm.pdf(x)
cdf = norm.cdf(x)

plt.plot(x, pdf, label="PDF")
plt.plot(x, cdf, label="CDF")
plt.title("PDF e CDF para Distribuição Normal")
plt.legend()
plt.show()
```

**12) Apresente o gráfico de probabilidade da função de massa de probabilidade (PMF) para a distribuição binomial (Probabilidade x Número de Sucessos).**

```python
from scipy.stats import binom

n = 10  # (Número de ensaios)
p = 0.5  # (Probabilidade de sucesso)

k = np.arange(n + 1)

pmf_values = binom.pmf(k, n, p)

plt.bar(k, pmf_values)
plt.xlabel('Número de Sucessos (k)')
plt.ylabel('Probabilidade')
plt.title('Gráfico de Probabilidade da PMF Binomial')
plt.show()
```

**13) Gere uma amostra aleatória e apresente a estatística t e o valor p**

```python
from scipy.stats import ttest_ind

amostra1 = np.random.normal (loc=0, scale=1, size=50) #Forma 1 de fazer
amostra2 = np.random.normal (0, 1, 50) #Forma 2 de fazer

statistic, p_value = ttest_ind(amostra1, amostra2)

print("Amostra 1:\n", amostra1)
print("\nAmostra 2:\n", amostra2)
print(f"\nEstatística t: {statistic}")
print(f"Valor p: {p_value}")
```

**14) Apresente os valores de correlação de Pearson e Spearman e um gráfico com linha de regressão ajustada sobre os pontos de dados.**

*a. A correlação de Pearson avalia a relação linear entre duas variáveis
contínuas.*

*b. A correlação de Spearman avalia a relação monotônica entre duas
variáveis contínuas ou ordinais.*

```python
from scipy.stats import pearsonr, spearmanr

x = np.random.rand(50)
y = 2.5 * x + np.random.normal(0, 0.2, 50)

pearson_coef, pearson_p = pearsonr(x, y)
print(f"A) Correlação de Pearson: {pearson_coef:.3f}, p-value: {pearson_p:.3f}")

spearman_coef, spearman_p = spearmanr(x, y)
print(f"B) Correlação de Spearman: {spearman_coef:.3f}, p-value: {spearman_p:.3f}")

slope, intercept = np.polyfit(x, y, 1)
regression_line = slope * x + intercept

plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Dados fornecidos')
plt.plot(x, regression_line, color='red', label='Linha de Regressão')
plt.xlabel('Variável X')
plt.ylabel('Variável Y')
plt.title('Gráfico de Dispersão com Linha de Regressão')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# A base deve ser tratada: existem valores nulos e dados categóricos
# Preencher valores nulos em colunas numéricas com a média
df.fillna(df.mean(numeric_only=True), inplace=True)

# Preencher valores nulos em colunas categóricas com a moda
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
# Separar o dataset em X (matriz de features) e y (coluna target)
target_column = "species"  # Coluna de destino (target)
X = df.drop(columns=[target_column])  # Features (todas as colunas, exceto a coluna target)
y = df[target_column]  # Rótulo (classe ou target)
```

```python
# Identificar colunas categóricas e numéricas
categorical_columns = X.select_dtypes(include=["object"]).columns  # Colunas categóricas
numerical_columns = X.select_dtypes(include=["number"]).columns  # Colunas numéricas
```

```python
# Normalizar todas as colunas (Valores das colunas entre 0 e 1)
# Criar um pipeline de pré-processamento para normalizar os numéricos e codificar os categóricos
preprocessor = ColumnTransformer(
    transformers=[
        ("num", MinMaxScaler(), numerical_columns),  # Normalizar numéricos entre 0 e 1
        ("cat", OneHotEncoder(), categorical_columns)  # Codificar categóricos como One-Hot
    ]
)

# Aplicar o pré-processamento nos dados
X_processed = preprocessor.fit_transform(X)
```

```python
# Gerar as bases de treinamento e teste (Informar porcentagem utilizada)
# Dividir os dados em 70% para treino e 30% para teste
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.3, random_state=42)
```

```python
# Treinar o modelo com os dados de treinamento (Informar porcentagem utilizada: 70% treinamento, 30% teste)
mlp.fit(X_train, y_train)

print("Formato de X_train:", X_train.shape)
print("Formato de y_train:", y_train.shape)
```

```python
# Fazer o predict com os dados de teste
y_pred = mlp.predict(X_test)
```
